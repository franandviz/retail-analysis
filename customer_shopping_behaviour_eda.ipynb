# Create Dimension Tables
Create dimensions for categorical fields with mutable or relevant values.
If possible, dates would be included to monitor changes/removals/additions using SCD type 2.

### Dimension Tables
* `item_purchased`
* `category`
* `location`
* `size`
* `color`
* `shipping_type`

```python
dim_col_dict = {'dim_item': 'item_purchased', 'dim_item_category': 'category', 'dim_us_location': 'location', 'dim_item_size': 'size', 'dim_item_color': 'color', 'dim_shipping_type':'shipping_type'}
```

```python
dim_df_dict = {}

for k,v in dim_col_dict.items():

    dim = list(set(df_comp[v]))
    id_name = k + '_id'
    dim_index = [i+1 for i in range(len(dim))]

    dim_df_dict[k] = pd.DataFrame({id_name: dim_index, v: dim})
```

```python
dim_df_dict
```

```python
for dim, df in dim_df_dict.items():
    df.to_csv(f'{file_location}\\{dim}.csv', index=False)
```
